Le vecteur est définit dans types.py
On utilise matplotlib pour le dessiner et numpy pour faire des calculs précis

```python
from rotation import vector
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import colors
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets
```

Commençons par tracer le vecteur V = \[1 2 3\] dans l'espace:

```python
# On crée le vecteur
V = vector([1, 2, 3])

# On prépare le graphe
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d", xlim = (-3, 3), ylim = (-3, 3), zlim = (-3, 3))
plt.ion()

# On dessine le vecteur
ax.quiver(0, 0, 0, V[0], V[1], V[2], color = "black")
plt.show()
```

Faisons-le tourner en x

```python
x_unitaire = vector([1, 0, 0])
def draw_rotation(v: vector, angle):
    global Q, fig, ax
    Vrot = v.rotate(x_unitaire, angle, True)
    ax.clear()
    ax.quiver(0, 0, 0, Vrot[0], Vrot[1], Vrot[2], color = "black")
    fig.canvas.draw()

angle_slider = widgets.IntSlider(min = 0, max = 360, step = 1)
widgets.interact(draw_rotation, v = widgets.fixed(V), angle = angle_slider)
plt.show()
```

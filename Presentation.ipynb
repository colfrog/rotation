Le vecteur est définit dans rotation.py
On utilise matplotlib pour le dessiner.

```python
%matplotlib notebook
from scene import *
from rotation import vector
import numpy as np
import ipywidgets as widgets
```

Commençons par tracer le vecteur V = \[1 2 3\] dans l'espace:

```python
# On crée une scène pour cette démonstration
s1 = scene1()

# On crée le vecteur V qui part de l'origine comme [2 2 2]
V = vector([2.0, 2.0, 2.0])

# On dessine le vecteur
s1.Q = s1.quiver(V)
s1.fig.show()
```

Si on pose un vecteur unitaire représentant l'axe des x, le vecteur V tournera selon le cercle à la base de ce vecteur, avec un rayon égal à la composante en x de v

```python
# Dessin du vecteur unitaire x à l'origine de V
s1.quiver(s1.xunit(V), "red")
```

Faisons-le tourner par rapport à l'axe des x

```python
# Production du slider qui définit l'angle du vecteur par rapport au vecteur unitaire en x
widgets.interact(s1.draw_2D_rotation, dir = widgets.fixed(s1.xunit(V)), \
                 v = widgets.fixed(V), \
                 angle = s1.make_angle_slider())
```

On peut rajouter des axes en y et z pour le faire tourner en 3D

```python
# On crée une deuxième scène pour la démonstration en 3D
s2 = scene2()

# On trace les vecteurs unitaires en x, en y et en z
s2.quiver(s2.xunit(V), "red")
s2.quiver(s2.yunit(V), "blue")
s2.quiver(s2.zunit(V), "green")

# On trace trois sliders pour définir l'angle autour des vecteurs unitaires en x, y et z
widgets.interact(s2.draw_3D_rotation,
                 v = widgets.fixed(V),
                 angleX = s2.make_angle_slider(),
                 angleY = s2.make_angle_slider(),
                 angleZ = s2.make_angle_slider())
```
